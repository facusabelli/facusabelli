```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df1 = pd.read_csv("viajesSube.csv")
df1 = df1.loc[0:200]
x_values = df1['TIPO_TRANSPORTE'].unique()
y_values = df1['TIPO_TRANSPORTE'].value_counts().tolist()
y_values.reverse()

# grfico 1
plt1.bar(x_values,y_values)
plt1.title('Transportes utilizados desde marzo 2020 hasta septiembre 2021')
plt1.xlabel('Tipo de transporte')
plt1.ylabel('Cantidad de personas')
plt1.show()

# grfico 2
plt2.plot(x_values,y_values);
plt2.title('Transportes utilizados desde marzo 2020 hasta septiembre 2021')
plt1.xlabel('Tipo de transporte')
plt1.ylabel('Cantidad de personas')
plt2.show()

```
