```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df1 = pd.read_csv("personal.csv")
df1 = df1.loc[0:500]
x_values = df1['Categoria'].unique()
y_values = df1['Categoria'].value_counts().tolist()
y_values.reverse()

# grfico 1
plt1.bar(x_values,y_values);
plt1.xticks(rotation=90,fontsize=14)
plt1.title('Puesto especializado')
plt1.xlabel('Tipo de transporte')
plt1.ylabel('Cantidad de personas')
plt1.show()

# grfico 2
plt2.plot(x_values,y_values);
plt1.xticks(rotation=90,fontsize=14)
plt2.title('Puesto especializado')
plt1.xlabel('Tipo de transporte')
plt1.ylabel('Cantidad de personas')
plt2.show()
```
